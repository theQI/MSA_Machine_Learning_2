```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X_df1, Y_df1, test_size=0.30,random_state=123) 

```

```python
from sklearn.ensemble import RandomForestClassifier  
clf = RandomForestClassifier() 
clf.fit(X_train, y_train)

y_predict = clf.predict(X_test)
print(confusion_matrix(y_test, y_predict)) 

print(accuracy_score(y_test, y_predict)*100) 
```

```python
# Use PCA to reduce features 
from sklearn.decomposition import PCA 
from sklearn.preprocessing import scale


p = scale(X_df1)

pca = PCA().fit(p)

pca = PCA(n_components=10)


pcafit = pca.fit_transform(p)

pcafitDf = pd.DataFrame(data = pcafit,columns = ['pc1', 'pc2', 'pc3', 'pc4', 'pc5', 'pc6', 'pc7', 'pc8', 'pc9', 'pc10'])
```

```python
p3 = pcafitDf


p3_train, p3_test, y_train3, y_test3 = train_test_split(p3, y, test_size=0.3, random_state=123)
```

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Pulsar Dataset Explained Variance')
plt.show()
```
